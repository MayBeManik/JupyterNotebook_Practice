# DATASET "EXERCISE"

```python
df=sns.load_dataset('exercise')
df.head()
```

# Dist Plot

```python
sns.displot(df.id, kde=False)
```

# Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Joint Plot

```python
sns.jointplot(x='pulse',y='id',data=df,kind='reg')
```

```python
sns.jointplot(x='pulse',y='id',data=df,kind='hex')
```

# Pair Plot

```python
sns.pairplot(df)
```

# Box Plot

```python
sns.boxplot(x='pulse',y="diet",hue='kind',data=df,orient="h",palette="rainbow")
```

# Bar Plot

```python
sns.barplot(y="pulse",x='diet',data=df,hue='kind')
```

# Count Plot

```python
sns.countplot(x='diet',data=df)
```

# Violin Plot

```python
sns.violinplot(x="pulse",y='diet',data=df,hue='kind')
```
