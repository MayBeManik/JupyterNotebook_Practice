```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sugar_men = [10,12,22,34,54,67,77,82,34,34,57,20,30,40,50]
sugar_women = [15,25,35,45,55,34,54,65,45,45,35,43,76,54,32]
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([sugar_men,sugar_women],bins=[10,30,50,70,90],rwidth = 0.95, color=['g','b'], label=['men','women'])
plt.legend()
```

```python
plt.hist(sugar_men,rwidth=0.95,bins=[10,30,50,70,90])
```

```python
plt.ylabel('sugar range')
plt.xlabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([sugar_men,sugar_women],bins=[10,30,50,70,90],rwidth = 0.95, color=['g','b'], label=['men','women'],
         orientation='horizontal')
plt.legend()
```
